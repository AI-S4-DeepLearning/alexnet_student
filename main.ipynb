## Stap 1
Importeren van de benodigde modules en het instellen van wat settings.

```python
import random
import torch
import numpy as np
from DatasetGenerator import DatasetGenerator
from AlexNetTrainer import AlexNetTrainer
from ResultPlotter import ResultPlotter

SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

## Stap 2
Inladen van data (er wordt nog niet gebalanceerd, zie generator.get_balanced_splits).

```python
generator = DatasetGenerator(SEED, data_dir='data', img_size=224)
print(f"Found {generator.num_classes} classes with {len(generator.full_dataset)} images: {generator.classes}.")
```

```python
#min_class_size staat nu op -1. Zodra je het balanceren van data hebt geÃ¯mplementeerd dient dit argument een logische waarde te krijgen.
train_ds, val_ds, test_ds = generator.get_balanced_splits(train_ratio=0.7, val_ratio=0.15, min_class_size=-1)
```

```python
print(f"After balancing {len(train_ds) + len(val_ds) + len(test_ds)} images.")
print(f" - Train: {len(train_ds)} samples (balanced + augmented)")
print(f" - Val:   {len(val_ds)} samples (no augmentation)")
print(f" - Test:  {len(test_ds)} samples (no augmentation)")
```

## Stap 3
Definieren van de juiste hyperparameters voor Gridsearch (dit is nog niet gedaan).

```python
# OPDRACHT: Kies goede parameterwaarden om grid search over uit te voeren.
param_grid = {
    'batch_size': [],
    'learning_rate': [],
    'dropout_rate': []
}
```

## Stap 4
Vinden van het beste model met Gridseach.

```python
epochs = 10
max_val_acc = 0
max_trainer = None

for lr in param_grid['learning_rate']:
    for bs in param_grid['batch_size']:
        for dropout in param_grid['dropout_rate']:
            print(f"\n{'='*100}")
            print(f"Training: learning rate={lr}, batch size={bs}, dropout={dropout}")
            print(f"{'='*100}")
            
            
            trainer = AlexNetTrainer(num_classes=generator.num_classes, 
                                    learning_rate=lr,
                                    batch_size=bs,
                                    dropout_rate=dropout,
                                    device=device)
            
            trainer.fit(train_ds, val_ds, bs, epochs)
            
            if trainer.history['val_acc'][-1] > max_val_acc:
                max_val_acc = trainer.history['val_acc'][-1]
                max_trainer = trainer
```
